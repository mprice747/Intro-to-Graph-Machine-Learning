# Baseline Model

Before we jump into modeling, let's come up with a simple baseline model. It’s a good habit to try something simple to get feel of the data, see any quirks, and gauge how difficult the problem is. In our case we will start with a simple method. Let's assume that movies that have similar links are related. For example if I liked movie A and movie A has the same director, same production company, same actors, .., to movie B then movie B could be a good recommendation for movie A. So for a given movie we will calculate the overlap with every other movie. The overlap will be the the number of elements in the intersection of the links between movie A and B divide by the number of links in the union of A and B. This procedure is called the Jaccard Distance. Notice this operation is symmetric, you could instead divide it by only the number of links in the set of movie A. That would make the operation asymmetric. Feel free to change the metric to anything you want and see the results. 

```python
def model_baseline(i):
    moviei_links = set(movies[i][WIKI_LINK])
    recommend = [0]*len(movies)
    for j in range(len(movies)):
        if j == i: continue
        moviej_links = movies[j][WIKI_LINK]
        recommend[j] = len(
            moviei_links.intersection(moviej_links))/len(moviei_links.union(moviei_links))
    return np.array(recommend)

def movie_recommend_baseline(i, num_rec=5, show_score=False):
    rec = model_baseline(i)
    indices = np.argsort(rec)[::-1]
    rec = rec[indices]
    
    j = seen = 0
    print(f'{movies[i][MOVIE_NAME]} recommendations\n'+'-'*40)
    while seen < num_rec:
        if indices[j] == i: j += 1
        else: 
            print(f'-{movies[indices[j]][MOVIE_NAME]}' + (', score={:.2f}'.format(rec[j]) if show_score else ''))
            seen += 1; j += 1
```

# Node2Vec 

Node2Vec is a popular unsupervised graph learning algorithm which transforms a graph's nodes into vector representations. With these vector representations, we can represent our graph as a traditional dataframe (or data matrix). Because most machine learning algorithms require data matrices as inputs, we will therefore transform our graph into a much more usuable format. <br>
<br>
In this part of the lab, we will implement a simple version of node2vec for our movie recommendation problem. However, to fully understand the algorithm, we will first review a number of prerequisites, namely basic probability, logisitic regression and word2vec.

## Bernoulli Distributions 

The Bernoulli distribution is one of the most basic probability distributions. Essentially, it models a binary outcome ($y$ either equal to 0 or 1), where a "1" happens with probability $\theta$ and a "0" happens with probability $1 - \theta$. Mathematically, it can be represented as <br> <br>
  $$
p(y|\theta) = (\theta)^{y}(1 -\theta)^{1 - y}
$$

Now, using this distribution on its own can only model incredibly simple situations, like flipping a coin. However, of course the probability of an outcome can change based on outside factors. Therefore, in order to to account for this, we need a more powerful model. 

```python
# Bernoulli Distribution
def bernoulli(y, theta):
    
    return (theta ** y) * (1 - theta)**(1 - y)
```

## Logistic Regression 

One of the most common methods to model how the probability of a binary outcome changes based on certain input data is logistic regression. In short, given some predictiors, $x = (x_{1}, x_{2}, ..., x_{p})$ we model the probability of a positive outcome as <br> <br>

$$
\theta(\beta, x) = \frac{1}{1 + e^{-(\beta_{0} + \beta_{1}x_{1} + ... + \beta_{p}x_{p})}}
$$

This is known as the sigmoid function, and it takes in any real number as input and outputs a number in between 0 and 1 exclusive, like a probability! A simplified version of the function ($\frac{1}{1 + e^{-x}}$) is plotted below.

```python
# Simple sigmoid function
def sigmoid(x):
    
    return 1/(1 + np.exp(-1 * x))

# Generate 1000 equally spaced numbers in between -10 and 10, and get outputs
x_axis = np.linspace(-10, 10, 1000)
y_axis = sigmoid(x_axis)

# Generate plot of sigmoid function
fig, ax = plt.subplots(figsize = (7, 5))
ax.plot(x_axis, y_axis, linewidth = 4, color = 'blue')
ax.set_title('Sigmoid Function', fontsize = 12)
ax.set_xlabel('Input', fontsize = 11)
ax.set_ylabel('Output', fontsize = 11)
plt.show()
```

Finally, it is common to represent both the coefficents and input data as vectors. That is $\beta = (\beta_{0}, \beta_{1}, ..., \beta_{p})$ and $x = (1, x_{1}, x_{2}, ..., x_{p})$. With this format, we can represent $\beta_{0} + \beta_{1}x_{1} + ... + \beta_{p}x_{p}$ as the dot product between $\beta$ and $x$ ($\beta \cdot x$). You'll notice that the sigmoid function is an increasing function, meaning that if $\beta \cdot x$ increases, the sigmoid increases as well, and thus the probability of whatever we are modeling. This will be important later!  <br><br>

Now, we have a model for the probability of a binary outcome given some predictors! Of course, the regression coefficents $\beta = (\beta_{0}, \beta_{1}, ..., \beta_{p})$ are unknown and we want to learn these parameters with our data. In machine learning, this is almost always achieved by optimizing some cost function. Now, what would be the cost function here?  <br> <br> 

Remember, we are modeling a binary outcome, so it seems like the Bernoulli distribution would come in handy. Indeed, all we need to do know is maximize the following function, given the outcomes $(y)$, the data matrix ($X$), and $\theta$ replaced with the sigmoid! <br>

$$
\arg \max_{\beta} f(\beta|X) \ =  \ \Pi_{i = 1}^{n} \theta(\beta, x_{i})^{y_{i}}(1 - \theta(\beta, x_{i}))^{1 - y_{i}} = \Pi_{i = 1}^{p}\theta(\beta, x_{i}) \Pi_{i = n - p}^{n}(1 - \theta(\beta, x_{i}))
$$

<br> 

Luckily, taking the natural log of this function yields something that can easily be optimizied with gradient descent (A maximation task can easily be turned into a minimization task by simply taking the negative of the cost function). We won't go over the derivative calculations, but this can be achieved automatically with [scipy's minimize function](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html) . 



```python
from scipy.optimize import minimize 
```

```python
# Theta(beta, x) function from above

def theta_beta_x(beta, x):
    
    return sigmoid(np.dot(beta, x))


def lr_cost_function(beta, X, y):
    '''
    Cost function for logisitic regression model
    
    Inputs: 
        beta - vector of length p representing the current estimate of beta
        X - a n x p data matrix
        y - n x 1 matrix of binary responses 
        
    Output: 
        cost_value - float representing value of cost function given the inputs
    '''
    
    # Initialize for the loop
    n = X.shape[0]
    rev_cost_value = 1
    
    # Because of log, the cost become a sum and the log is distributed
    for i in range(n):
        
        new_theta = theta_beta_x(beta, X[i, :])
        
        rev_cost_value += np.log(bernoulli(y[i, 0], new_theta))
        
    
    # Turn into negative because scipy only supports minimization procedures
    cost_value = -1 * rev_cost_value
        
    return cost_value
        

def fit_logistic_regression(X, y):
    '''
    Fits a standard logistic regression model with scipy's minimize function
    
    Inputs: 
        X - a n x p data matrix
        y - n x 1 matrix of binary responses 
        
    Output:
        result - output from scipy minimize function given X, y and the 
        logistic regression cost function
    '''
    
    # Initial guess for beta
    initial_beta = np.random.normal(size = X.shape[1])
    
    # Function for scipy to minimize, with beta as input and X and y
    # as constant
    for_scipy = lambda beta : lr_cost_function(beta, X, y)
    
    # Get beta estimates from scipy, as well as other diagnostics 
    result = minimize(for_scipy, x0 = initial_beta, method = 'BFGS', 
                      tol = 1e-6)
    
    return result
```

Now, to prove our code works we will generate some fake data from a known model, and see if our estimates are reasonably close to the ground truth. We will assume the true beta coefficents are $ \beta = (1.1, -0.7, -2.5, 3.4)$

```python
# Real Betas
real_betas = np.array([1.1, -0.7, -2.5, 3.4])

# Set seed and dimensions
np.random.seed(20)
n = 1000
dim = real_betas.shape[0] - 1

# Generate fake data matrix (simulated from multivariate normal) 
# and add column full of 1's for intercept
fake_X = np.random.multivariate_normal(mean = np.array([0] * dim),
                                       cov = np.eye(dim), size = n)
fake_X_1 = np.concatenate([np.array([1] * n).reshape(-1, 1), fake_X], axis = 1)

# Get actual probabilities and simulate from obtained 
# joint bernoulli distribution
real_probabilities = sigmoid(np.dot(fake_X_1, real_betas.reshape(-1, 1)))
fake_y = np.array([np.random.binomial(1, x[0]) for x in real_probabilities])
fake_y = fake_y.reshape(-1, 1)

# Fit the data and get estimates for beta
fitted_betas = fit_logistic_regression(fake_X_1, fake_y)['x']

# Turn arrays into strings
s_real_betas = '(' + ' '.join([str(x) + ',' for x in real_betas])[:-1] + ')'
s_fit_betas = '(' + ' '.join([str(x) + ',' for x in fitted_betas])[:-1] + ')'


print('Real Betas: ' + s_real_betas)
print('Estimated Betas: ' + s_fit_betas)
```

Just as we expected, we have obtained reasonable estimated of the ground truth. There are two things we want you to take away fom this. One, in a learned logistic regression model, as the dot product between the beta vector and the input vector increases, the probability ($\theta$) also increases. In addition, the logistic regression coefficients are estimated by maximizing the natural log of the joint probability distribution. <br> <br>

## Word2Vec

Natural Language Processing is a popular research area in machine learning. Because almost all machine learning procedures are numerical algorithms, they do not take raw text as input. Therefore, raw text is almost always transformed into a real numbered vector, called a word embedding. Ideally, words that are similar to each other are "close" to each other in the vector space. word_embedding_img.png

Word2Vec is an unsupervised word embedding technique developed by Google in 2013. The procedure is based on the distributional hypothesis in linguistics, which states that words that appear "near" each other tend to have similar meanings. For example, because the words "player" and "coach" are likely to appear in the same sentence or a similar context, they are much more likely to be related compared to "player" and "pilot." 

Now, how to we define "near"? For each word, we'll define a window, and if another word falls into that window, they are considered related. The length of this window is arbitrary, and can be considered a hyperparameter that needs to be tuned. 

dist_hyp.png

In this example, we have a window size of 2, and any words outside of this are considered unrelated.

Now, how can model this idea numerically and probabilistically? Consider a target word ($w$) and a potential "context" word ($c$). Essentially, we are interested if the words are related or not. This sounds like a binary classification task, so naturally, we can attempt to model the probability of these words being related. Put mathematically, <br>
$$
p(related| w, c)
$$

Remember, we want to transform $w$ and $c$ into vectors. Remember from earlier in logistic regression, as the dot product between $\beta$ and $x$ increased the probability of the event of interest also increased? We can use this same idea here, aka as the dot product between $w$ and $c$ increases, the probability of being related increases as well! Or put mathematically, <br>

$$
p(related|w, c) = \theta(c, w) = \frac{1}{1 + e^{-(c \cdot w)}}
$$

## Fitting Word2Vec

Of course, in order to fit this model, we need examples of both related and unrelated words. For each word ($w$), we can gather all the context words that are in the predefined windows. These will be our positive examples ($c_{pos}$). Then, we can randomly sample words from the corpus. These will be our negative examples ($c_{neg}$). Because this can be seen as a binary classification task, we can use the same cost function from logistic regression! Put mathematically:  <br> <br>

$$
Original: \ \ \Pi_{i = 1}^{p}\theta(\beta, x_{i}) \Pi_{i = n - p}^{n}(1 - \theta(\beta, x_{i})) \\ 
$$

$$
New: \ \ \Pi_{j = 1}^{n}[\Pi_{i = 1}^{p_{j}} \theta(c_{i,pos}, w_{j})\Pi_{i = n_{j} - p_{j}}^{n_{j}}(1 - \theta(c_{i,neg}, w_{j}))]
$$

This again be optimized with gradient descent. It is worth noting that every word will appear as both a "target" word and a "context" word. These are considered two different embeddings. Therefore, these are often combined, either by element wise multiplication or addition, to get one embedding per word after training. <br> <br> 

We won't be coding up this model from scratch, as this is a much harder optimization problem. Instead we will be using a library called gensim to perform all our calculations. We will again generate some fake data, and assume our text corpus consists of 100 words, numbered 1 from 100. We will also generate it in such a way such that a number is more more likely to appear near its "neighbors." <br> <br> 

Gensim's w2v model takes in a list of lists, with the lists representing "sentences" and each individual element representing a word. We will be generating 20000 sentences with 20 words using the sampling strategy described above. 

```python
np.random.seed(20)

# Set vocab size and intialize word list
vocab_size = 100
word_list_str = [str(x) for x in range(vocab_size)]

# Set number of sentences, words per sentence and sampling parameter. 
num_sentences = 20000
words_per_sentence = 20
sampling_param = 4 # This is the alpha and beta parameter from a beta distribution

# Final corpus 
final_corpus = []

# Samples sentences
for i in range(num_sentences):
    
    # Sample a starting word, and sample from a beta distribution
    # with range(-vocab_size/2, vocab_size/2)
    starting_word = np.random.choice(word_list_str)
    sentence = [starting_word]
    random_gen_sent = (np.random.beta(a = sampling_param, b = sampling_param, 
                  size = words_per_sentence) * vocab_size) - vocab_size/2  
    
    # Round results and turn into integers 
    random_gen_sent_int = np.round(random_gen_sent).astype(int)
    
    # Generate sentence by adding sampled value to previous result
    for j in random_gen_sent_int: 
        
        word_num = (int(sentence[-1])+ j) % vocab_size

        sentence.append(str(word_num))
        
    final_corpus.append(sentence)
    
print("Sentence Example:", final_corpus[0])
```

Now with our corpus generated, we can finally train our word2vec model! We have set the vector size to 20, and then to visualize our results, we will perform Principal Components Analysis to see if the results make sense.

## Transforming Data to a Graph


We could frame the problem as the following. We have a list of movies that interact with a list of wikipedia pages. We want to find the following relationships: movie to movie, movie to link, and link to link. This is similar to an e-commerce store wanting to know which users are similar to other users, which products are users more likely to buy, what products are purchased together. <br> <br> 

From this point of view our data is a Bipartite graph. A Bipartite graph is a graph that has two sets of nodes where the connections are only between  different node types. We can model one set of nodes as movies and the second set as wiki-links. The idea is to build a representation of the graph, such that each node embeds its neighborhood information.
bipartite.png

```python
# Get links that occur more than 3 times but less than 6000
links = [link for link, c in link_counts.items() if c >= 3 and c<=6000]

# Map a number identifer to each link and movie
link_to_index  = {link: index for index, link in enumerate(links)}
movie_to_index = {movie[MOVIE_NAME]: index for index, movie in enumerate(movies)}

# If a movie page has a certain wikipedia link, add tuple containing the movie and link identifiers
# These represent the edges of our graph
movie_link = []
for movie in movies:
    movie_link.extend(
        (movie_to_index[movie[MOVIE_NAME]], link_to_index[link])
        for link in movie[WIKI_LINK] if link in link_to_index)
```

```python
# Get edges, movie nodes and link nodes
edges = list(set([('m_'+str(o[0]), 'l_'+str(o[1])) for o in movie_link]))
movie_nodes = list(set([o[0] for o in edges]))
link_nodes  = list(set([o[1] for o in edges]))
print("Number of movie nodes {}, Number of link nodes {}".format(len(movie_nodes), len(link_nodes)))
```

To represent our graph data, we will be using the graph object from the [networkx library](https://networkx.org/documentation/stable/index.html). 

```python
# Python graph library. Graph ML functions typically take a nx graph object as input
import networkx as nx
```

```python
# Make bipartite graph
graph = nx.Graph()
graph.add_nodes_from(movie_nodes, bipartite=0, label='movie')
graph.add_nodes_from(link_nodes, bipartite=1, label='link')
graph.add_edges_from(edges, label='wiki')

# Print Graph info
print(nx.info(graph))
```

## Fitting Node2Vec

One way to think about transferring the ideas from Word2Vec to graphs is for each node, consider its neighbors as related and anything outside that neighborhood as unrelated. This is on the right track but there are two issues with this startegy. First, in a large text corpus, most words will appear multiple times. In a graph, however, each node only appears once, giving us very limited to data to train on. The second issue can be explained visually. <br> 

graph_ex.png


Let's look at node 3. We can see it is part of a "community" with nodes 4 through 8. However, assume we define our neighborhood "window" to be 2. Therefore, because node 1 and nodes 5 and 6 are a distance of 2 away, to node 3, they are equally related. However, nodes 5 and 6 are part of the same community as node 3, while node 1 is not. Therefore, to better capture the communities within graphs, we need a different approach. <br> <br> 

Remember Word2Vec takes in a list of lists, which are supposed to be "sentences". So how do we generate "sentences" in graphs? We will do so by generating random walks. Essentially, we first have a node as a "starting point". Then, we randomly pick one of its neighbors, and travel to that node. Then we randomly pick one of that node's neighbors and the cycle repeats. This process continues until we have a "sentence" or list of nodes. This same algorithm is repeated with each node being a starting point multiple times. Now we can run Word2Vec on this list of lists! <br> <br> 

Now how does this strategy solve our two issues? First each node will be a starting point and most likely is connected to other nodes in the graph. Therefore, it will likely show up multiple times in the output, solving the limited data problem. In addition, the random walk is most likely going to travel through the "community" rather than to unrelated points. Therefore, in the output, nodes in the same community are almost certan to appear near each other in the final output. 

```python
# Check if there are any isolated nodes by returning all connected components
training_graph = list((graph.subgraph(c) for c in nx.connected_components(graph)))[0]
print(nx.info(training_graph))
```

## Building a Node2Vec Movie Recommender

Now that we have our movies and links represented as vectors, it's time to build our movie recommendation system. Given a movie, we want to find the movies that are the most similar. Because each movie is now represented as a vector, we will use the cosine simialrity, essentially a normalized dot product. This metric ranges from -1 to 1, with higher values representing higher similarity.

```python
def dot_norm(u, v):
    # Normalized dot product (cosine similarity)
    return (u*v).sum()/np.sqrt( (u**2).sum()*(v**2).sum() )

def movie_recommend_node2vec(i, num_rec=5, show_score=False):
    
    '''
    Given a movie index, prints the top num_rec recommended movies according to node2vec
    
    i - movie index
    num_rec - number of movies to recommend
    show_score - whether to show cosine similarity score or not 
    '''
    
    # List of tuples of length 3, containing movie index, node indentifer, and its cosine similarity
    res = []
    u = model_node2vec.wv['m_'+str(i)]
    
    # Get movie cosine similarity score for all movies
    for node in list(model_node2vec.wv.vocab):
        if 'm' in node:
            v = model_node2vec.wv[node]
            res.append(('m_'+str(i), node, dot_norm(u, v)))
            
    j = seen = 0
    # Sort by the cosine similarity in descending order
    rec = sorted(res, key=lambda o: o[2], reverse=True)
    
    # Print results
    print(f'{movies[i][MOVIE_NAME]} recommendations\n'+'-'*40)
    while seen < num_rec:
        pre = rec[j]
        if pre[1] == 'm_'+str(i): pass
        else:
            print(f'-{movies[int(pre[1][2:])][MOVIE_NAME]}' + (', score={:.2f}'.format(pre[2]) if show_score else ''))
            seen += 1
        j += 1
```

## Building a Recommender with Deep Learning (PyTorch)

```python
import torch
from torch import nn
from torch.nn import functional
from torch.utils.data import Dataset, DataLoader
from tqdm.notebook import tqdm
```

Lastly, we can model the problem using pytorch. Similar to the graph model, we want to have model that builds an embedding for each movie and each wiki-link. Instead of generating random walks, we can first set each movie and link to random vectors. Then, we can sample random movie and wiki-link pairs and see if they are connected. Finally, we want to modify the vectors such that real pairs have a cosine similairty near one and fake pairs have a cosine similairity near -1. 

Another way to view this problem is that we are learning a metric. Given a movie index and a link index the metric will tell you how similar or dissimilar they are. Either way, this can be formulated as a deep learning problem, and we will build this neural network with PyTorch.

First let’s create a data set that returns a true pair with probability `positive_prob` and a false pair with probability `1-positive_prob`. In the positive case we sample from the true connections and in the false case we will randomly sample a movie and then a wiki-link such that is not linked. The return value from the data set will be tuple `((movie_id, link_id), label)`, where `label=1 ` is a true connection and `label=-1` is a false connection.

```python
class MovieLinkDataset:
  def __init__(self, movie_link, links, positive_prob=0.2): 

    '''
    Movie dataset class for PyTorch
    
    movie_link - a list of length 2 tuples. The first element of each tuple is the movie number identifier
    and the second element is the link number identifier.
    links - list of link names
    positive_prob - real number between 0 and 1. When sampling the dataset, the percent chance to sample
    a 'real' movie-link pearl is positive_prob, while sampling a 'fake' one is 1 - positive_pron

    '''

    # Build fields for class
    self.links = links
    self.positive_prob = positive_prob
    self.movie_link = list(set(movie_link))
    self.set_movie_link = set(movie_link)

    self.num_movies = len(set([o[0] for o in movie_link]))
    self.num_links  = len(set([o[1] for o in movie_link]))
      
  def __len__(self):

    # Return length of movie_link
    return len(self.movie_link)

  def __getitem__(self, idx):

  # Sample a real movie-link pair with probability positve_prob
    if random.random() < self.positive_prob:
      idx = random.randrange(len(self.movie_link))
      return self.movie_link[idx], 1
      
    while True:
      # Sample a fake movie-link pair with probability 1 - positive_prob
      movie_id = random.randrange(self.num_movies)
      link_id  = random.randrange(self.num_links)
      if not (movie_id, link_id) in self.set_movie_link:
          return (movie_id, link_id), -1
```

Our model will take the movie index and link index and retrieve their vector representation and perform the cosine similarity operation. We will use MSE loss such that back propagation will move the vectors such that true pairs will approach `1` and false pairs will approach `-1`.  Feel free to experiment by modifying the sampling strategy, adding additional layers, or changing the loss function.

```python
# Given a movie and link index

class Rank(nn.Module):

  '''
  Neural network architecture for movie and link embeddings

  num_movie - movie number identifer 
  num_link - lin number identifier
  emd_dim - embedding dimension for neural network
  '''
  def __init__(self, num_movies, num_links, emb_dim=64):

      # Initialize the embedding layers and the output layer
      super().__init__()
      self.movie_embedding = nn.Embedding(num_movies, emb_dim)
      self.link_embedding  = nn.Embedding(num_links,  emb_dim)
      self.similar = nn.CosineSimilarity()
      
  def forward(self, movie_indices, link_indices):

    # Forward pass for neural networ
      return self.similar(
          self.movie_embedding(movie_indices),
          self.link_embedding(link_indices))
```

```python
# Initialize data loader for training. Samples batch sizes of 4096, with about 0.2 of that
# being 'real' movie-link pairs and 0.8 being 'fake' movie-link pairs
ml = MovieLinkDataset(movie_link, links, positive_prob=0.2)
dl = DataLoader(ml, batch_size=4096)

# Initialize model, loss function and optimizer
model = Rank(ml.num_movies, ml.num_links)
loss_func = nn.MSELoss(reduction='sum')
optim = torch.optim.Adam(model.parameters(), lr=5.0e-2)
```

```python
# Number of training epochs (Number of iterations for training)
num_epochs = 15
global_step = 0

for epoch in range(num_epochs):
    print(f"Epoch: {epoch}")
    running_loss = 0
    running_size = 0

    # Sample batch from dataset
    for batch in tqdm(dl):
        (movie_ids, link_ids), labels = batch

        labels = labels.to(torch.float32)
        
        # Forward Pass
        outputs = model(movie_ids, link_ids)
        loss = loss_func(outputs, labels)
        
        # Print Out Metrics
        batch_size = movie_ids.shape[0]
        running_size += batch_size
        running_loss += loss.item()
        
        # Update
        optim.zero_grad()
        loss.backward()
        optim.step()
        global_step += 1
   
    print("Loss after {} steps: Running MSE={:.6f} Last Batch MSE={:.6f}".format(
        global_step, running_loss/running_size, loss.item()/batch_size))
```

```python

def movie_recommend_pytorch(i, num_rec=5, show_score=False):

  '''
  Movie Recommeder using learnt Pytorch weights

  Inputs: 
    i - movie index 
    num_rec - number of recommendations to print our
    show_score - boolean on whether to show similarity score

    Outputs: 

      Prints out the num_recth most similar movies according to Pytorch model

  '''

    # Get embeddings weights for each movie and sort by similarity score
  rec = functional.cosine_similarity(model.movie_embedding.weight[i], model.movie_embedding.weight).detach().numpy()
  indices = np.argsort(rec)[::-1]
  rec = rec[indices]

  # Print out results
  j = seen = 0
  print(f'{movies[i][MOVIE_NAME]} recommendations\n'+'-'*40)
  while seen <= num_rec:
      if indices[j] == i: j += 1
      else: print(f'-{movies[indices[j]][MOVIE_NAME]}' + (', score={:.2f}'.format(rec[j]) if show_score else ''))
      seen += 1; j += 1
```
